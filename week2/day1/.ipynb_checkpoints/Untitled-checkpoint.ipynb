```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import re
import statsmodels.api as sm
from scipy.stats import boxcox

pd.options.display.max_rows = 100
pattern = re.compile(r"^(?:\\.|[^/\\])*/((?:\\.|[^/\\])*)/")

def load_original_data():
    file1 = pd.read_csv('Data/file1.csv')
    file2 = pd.read_csv('Data/file2.csv')
    file3 = pd.read_csv('Data/file3.csv')
    return pd.concat([file1,file2,file3], axis=0)

ic_df=load_original_data()
ic_df.columns = list(map(lambda h:h.lower().strip(),ic_df.columns))
ic_df = ic_df.drop_duplicates(subset = "customer")

def string_to_float_int_convertor(data):
    if isinstance(data, str):
        data = data.replace('%','')
        return int(float(data))
    else:
        return data

ic_df['customer lifetime value'] = ic_df['customer lifetime value'].fillna(0)
ic_df['customer lifetime value'] = list(map(string_to_float_int_convertor,ic_df['customer lifetime value']))
ic_df['number of open complaints'] = ic_df['number of open complaints'].fillna(0)

def get_value_between_slashes(open_contract):
    if  isinstance(open_contract, str):
        return pattern.match(open_contract).group(1)
    else:
        return open_contract

ic_df['number of open complaints'] = list(map(get_value_between_slashes,ic_df['number of open complaints']))
ic_df['st'] = ic_df['st'].fillna('not provided')
ic_df['gender'] = ic_df['gender'].fillna('not available')

ic_df
```
